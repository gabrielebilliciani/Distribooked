```python
import json
import gzip
import os

# Define the file paths
input_file = "datasets/meta_Books.jsonl.gz"  # Adjust the path to your file
output_folder = "datasets/original_chunks/"  # Folder where chunks will be saved

# Define the chunk size (number of records per file)
chunk_size = 100000  # Adjust based on your needs


# Read the gzip JSONL file and split into chunks
with gzip.open(input_file, "rt", encoding="utf-8") as infile:
    chunk = []
    chunk_count = 0

    for line_number, line in enumerate(infile, 1):
        record = json.loads(line.strip())
        chunk.append(record)

        # When chunk size is reached, save to a new file
        if len(chunk) >= chunk_size:
            output_file = os.path.join(output_folder, f"chunk_{chunk_count:04d}.jsonl")
            with open(output_file, "w", encoding="utf-8") as outfile:
                for record in chunk:
                    outfile.write(json.dumps(record) + "\n")
            print(f"Chunk {chunk_count} saved with {len(chunk)} records.")
            chunk_count += 1
            chunk = []  # Reset the chunk

    # Save any remaining records as the final chunk
    if chunk:
        output_file = os.path.join(output_folder, f"chunk_{chunk_count:04d}.jsonl")
        with open(output_file, "w", encoding="utf-8") as outfile:
            for record in chunk:
                outfile.write(json.dumps(record) + "\n")
        print(f"Final chunk {chunk_count} saved with {len(chunk)} records.")
```
